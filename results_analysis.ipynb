```python
champs = []
for i in range(10_000):
   df = pd.read_csv(rf'brackets\tournament_results{i}.csv')
   champs.append(df[['national champion']])
```

```python
df_champs = pd.concat(champs)
pd.DataFrame(df_champs.value_counts()/10000).head(20)
```
